```python
import os

import pandas as pd
from IPython.display import display

# INPUT_PATH = 'car-price-prediction/data/carsWithImages.csv'
# OUTPUT_PATH = 'car-price-prediction/data/carsWithImageCleaned.csv'
INPUT_PATH = os.path.join("..", "data", "carsWithImages.csv")
OUTPUT_PATH = os.path.join("..", "data", "carsWithImageCleaned.csv")

df = pd.read_csv(INPUT_PATH, index_col=0)
```

```python
df.head()
```

```python
# remove electric cars

df = df[df['Combustibil'] != 'Electric']
```

```python
# see unique values in column Combustibil

df.Combustibil.unique()
```

```python
# delete Benzina + CNG and Benzina + GPL

df = df[df['Combustibil'] != 'Benzina + CNG']
df = df[df['Combustibil'] != 'Benzina + GPL']
```

```python
# see unique values in column Transmisie

df.Transmisie.unique()
```

```python
# replace 4x4 (automat) and 4x4 (manual) with 4x4

df['Transmisie'] = df['Transmisie'].replace(['4x4 (automat)'], '4x4')
df['Transmisie'] = df['Transmisie'].replace(['4x4 (manual)'], '4x4')

# rename column Transmisie to Tractiune

df.rename(columns={'Transmisie':'Tractiune'})
```

```python
# see RON and EUR counts

df['Pret'].apply(lambda s: s[-3:]).value_counts()
```

```python
# few RON => erase RON cars

df = df[df['Pret'].apply(lambda s: s[-3:]) != 'RON']

# erase EUR from prices

df['Pret'] = df['Pret'].map(lambda x: x.rstrip(' EUR'))

# rename column to include (EUR)

df = df.rename(columns={'Pret':'Pret (EUR)'})
```

```python
# remove space from numbers

df['Pret (EUR)'] = df['Pret (EUR)'].str.replace(" ", "")
df['Pret (EUR)'] = df['Pret (EUR)'].str.replace(",", ".") # Some prices contain a comma (Romanian locale).
df['Pret (EUR)'] = df['Pret (EUR)'].astype(float)
```

```python
# drop columns

df = df.drop(columns=['Tara de origine', 'Garantie dealer (inclusa in pret)', 'Data primei inmatriculari', 'Are VIN (Serie sasiu)', 'Numar locuri', 'Generatie', 'Optiuni culoare', 'Versiune'])
```
